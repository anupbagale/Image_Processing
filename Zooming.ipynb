```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('/content/drive/MyDrive/image_processing_lab/leena_img.png')
```

## Zoom by Replication method

```python
def zoom_in_replication(image, zoom_factor):
    height, width, channels = image.shape

    new_height = int(height * zoom_factor)
    new_width = int(width * zoom_factor)

    zoomed_image = np.zeros((new_height, new_width, channels), dtype=image.dtype)

    for i in range(new_height):
        for j in range(new_width):
            orig_i = int(i / zoom_factor)
            orig_j = int(j / zoom_factor)
            zoomed_image[i, j] = image[orig_i, orig_j]

    return zoomed_image
```

```python
zoom_factor = 10

zoomed_image = zoom_in_replication(image, zoom_factor)
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
zoomed_image_gray = cv2.cvtColor(zoomed_image, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image_gray, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Zoomed Image (Replication)')
plt.imshow(zoomed_image_gray, cmap='gray')

plt.show()
```

## Bilinear interpolation

```python
zoom_factor = 2

height, width = image.shape[:2]

new_height = int(height * zoom_factor)
new_width = int(width * zoom_factor)

zoomed_image = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_LINEAR)

image= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
zoomed_image = cv2.cvtColor(zoomed_image, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(12, 6))
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Zoomed Image (Bilinear Interpolation)')
plt.imshow(zoomed_image, cmap='gray')

plt.show()
```
