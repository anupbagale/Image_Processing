```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path= "/content/drive/MyDrive/image_processing_lab/leena_img.png"
```

```python
def contrast_stretching(image):
    min_pixel = np.min(image)
    max_pixel = np.max(image)

    # Apply the contrast stretching formula
    stretched_image = (image - min_pixel) * 255.0 / (max_pixel - min_pixel)

    # Convert to unsigned 8-bit integer type
   # stretched_image = np.clip(stretched_image, 0, 255)  # Ensure values are within [0, 255]
    stretched_image = stretched_image.astype(np.uint8)

    return stretched_image
```

```python
image= cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
stretched_image=contrast_stretching(image)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(stretched_image, cmap='gray')
plt.title('Stretched Image')
plt.show()

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.hist(image.flatten(), bins=256, range=(0, 256))

plt.subplot(1,2,2)
plt.hist(stretched_image.flatten(), bins=256, range=(0, 256))
plt.show()
```
