```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image= cv2.imread('/content/drive/MyDrive/image_processing_lab/leena_img.png',0)
```

## Sobel operator

```python
hor_filter = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
ver_filter = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])

# Apply Sobel filters
sobelX = cv2.filter2D(image, cv2.CV_64F, hor_filter)
sobelY = cv2.filter2D(image, cv2.CV_64F, ver_filter)

# Compute gradient magnitude
sobel_combined = np.sqrt(sobelX**2 + sobelY**2)

# Normalize to range [0, 255] and convert to uint8
sobel_combined = np.clip(sobel_combined, 0, 255).astype(np.uint8)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Sobel X')
plt.imshow(np.clip(sobelX, 0, 255).astype(np.uint8), cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Sobel Combined')
plt.imshow(sobel_combined, cmap='gray')
plt.axis('off')

plt.show()
```

## Prewitt operator

```python
hor_filter = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
ver_filter = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])

# Apply Prewitt filters
prewittX = cv2.filter2D(image, cv2.CV_64F, hor_filter)
prewittY = cv2.filter2D(image, cv2.CV_64F, ver_filter)

# Compute gradient magnitude
prewitt_combined = np.sqrt(prewittX**2 + prewittY**2)

# Normalize to range [0, 255] and convert to uint8
prewitt_combined = np.clip(prewitt_combined, 0, 255).astype(np.uint8)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Prewitt X')
plt.imshow(np.clip(prewittX, 0, 255).astype(np.uint8), cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Prewitt Combined')
plt.imshow(prewitt_combined, cmap='gray')
plt.axis('off')

plt.show()
```
