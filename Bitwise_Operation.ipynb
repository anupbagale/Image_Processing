```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

## Import image

```python
black_image=cv2.imread('/content/drive/MyDrive/image_processing_lab/black_background_wite_circle.png', cv2.IMREAD_GRAYSCALE)
white_image= cv2.imread('/content/drive/MyDrive/image_processing_lab/full_white.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
if black_image is None:
  print("black_image is None")
if white_image is None:
  print("white_image is None")
```

```python
plt.subplot(1,2,1)
plt.imshow(black_image, cmap='gray')

plt.subplot(1,2,2)
plt.imshow(white_image, cmap='gray')
plt.show()
```

## Check size of the image
 Since size are different, resizing it to the same size

```python
print(black_image.shape)
print(white_image.shape)
```

```python
black_image=cv2.resize(black_image, (256,256))
white_image=cv2.resize(white_image, (256,256))
```

```python
print(black_image.shape)
print(white_image.shape)
```

```python
black_img_data= np.array(black_image)
white_img_data= np.array(white_image)
```

```python
print(white_img_data)
```

# **Bitwise OR**

At first, create the new array with same size of black_image i.e 256*256. After that perform the bitwise OR operation between each corresponding pixels of black and white image. Place the output of the each operation in new array.

```python
result= np.zeros_like(black_image)
for i in range(black_img_data.shape[0]):
  for j in range(black_img_data.shape[1]):
    if white_img_data[i,j] == 255 and black_img_data[i,j] ==255:
       result[i,j] = 255
    elif white_img_data[i,j] == 255 and black_img_data[i,j] ==0:
      result[i,j] = 255

    elif white_img_data[i,j] == 0 and black_img_data[i,j] ==255:
      result[i,j] = 255

    else:
      result[i,j] = 0

output_image = result.astype(np.uint8)
plt.imshow(output_image, cmap='gray')
plt.title("OR using for loop")
plt.show()
```

## **Bitwise AND**

```python
result_and= np.zeros_like(black_image)
for i in range(black_img_data.shape[0]):
  for j in range(black_img_data.shape[1]):
    if white_img_data[i,j] == 255 and black_img_data[i,j] ==255:
       result_and[i,j] = 255
    else:
      result_and[i,j] = 0

output_image = result_and.astype(np.uint8)
plt.imshow(output_image, cmap='gray')
plt.title("AND using for loop")
plt.show()
```

## **Bitwise XOR**

```python
result_xor= np.zeros_like(black_image)
for i in range(black_img_data.shape[0]):
  for j in range(black_img_data.shape[1]):
    if white_img_data[i,j] == 255 and black_img_data[i,j] ==0:
       result_xor[i,j] = 255

    elif white_img_data[i,j] == 0 and black_img_data[i,j] ==255:
      result_xor[i,j] = 255

    else:
      result_xor[i,j] = 0

output_image = result_xor.astype(np.uint8)
plt.imshow(output_image, cmap='gray')
plt.title("XOR using for loop")
plt.show()
```

## **Bitwise Operation using function**

```python
image_or = cv2.bitwise_or(black_image, white_image)
image_and = cv2.bitwise_and(black_image, white_image)
image_xor = cv2.bitwise_xor(black_image, white_image)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(image_or, cmap='gray')
plt.title("OR")

plt.subplot(1,3,2)
plt.imshow(image_and, cmap='gray')
plt.title("AND")

plt.subplot(1,3,3)
plt.imshow(image_xor, cmap='gray')
plt.title("XOR")

plt.show()
```
